## separacion serie de tiempo

```python
def split(serie, trainportion):
    perct = math.ceil(len(ts) * trainportion)
    perc2 = math.ceil(len(ts)*((1 - trainportion)/2))
    train = serie[0:perct]
    validation = serie[perct:perct+perc2]
    test = serie[perct+perc2:]
    return train, validation, test
```

## suavizacion primer orden

```python
#hacer el modelo
def firstsmooth(y, lambda_):
    ytilde = y.copy()
    start = y[0]
    ytilde[0] = lambda_ * y[0] + (1 - lambda_) * start
    for i in range(1, len(y)):
        ytilde[i] = lambda_ * y[i] + (1 - lambda_) * ytilde[i - 1]
    return ytilde
```

```python
#pronosticar con primer orden
def forecastfirstsmooth(ts, actual, forecast, lambda_):
    T = len(actual) #longitud del training 129
    tau = len(forecast) # tiempo que se va a pronosticar 28
    #alpha_lev = 0.05 #significancia para los intervalos de prediccion
    cpi_forecast = np.zeros(tau)
    #cl = np.zeros(tau)
    cpi_smooth1 = np.zeros(T + tau)
    cpi_dates = ts

    for i in range(1, tau + 1): #esto va de 1 a 28
        cpi_smooth1[:T + i - 1] = firstsmooth(y=cpi_dates.iloc[:T + i - 1], lambda_=lambda_)
        cpi_forecast[i - 1] = cpi_smooth1[T + i - 2] + lambda_*(cpi_dates[T + i - 2] - cpi_smooth1[T + i- 2]) #posicion cero
        #sig_est = np.sqrt(np.var(cpi_dates.iloc[1:T + i - 1] - cpi_hat[:-1]))
        #cl[i - 1] = norm.ppf(1 - alpha_lev / 2) * sig_est
    return cpi_forecast
```

```python
#escoger el mejor modelo
best_mape=900
results=[]
for lambda_ in np.arange(0, 0.8, 0.05):
    cpi_forecast=forecastfirstsmooth(ts, train, validation, lambda_=lambda_)
    mape = 100*(sum(abs(((validation - cpi_forecast)/validation)))/len(validation))
    if  mape<best_mape:
        best_mape = mape
        ls = lambda_
    results.append(mape)
```

```python
#pronosticar con segundo orden
def forecastsecondsmooth(ts, actual, forecast, lambda_):
    T = len(actual) #longitud del training
    tau = len(forecast) # tiempo que se va a pronosticar
    alpha_lev = 0.05 #significancia para los intervalos de prediccion
    cpi_forecast = np.zeros(tau) 
    cl = np.zeros(tau)
    cpi_smooth1 = np.zeros(T + tau)
    cpi_smooth2 = np.zeros(T + tau)
    cpi_dates = ts

    for i in range(1, tau + 1):
        cpi_smooth1[:T + i - 1] = firstsmooth(y=cpi_dates.iloc[:T + i - 1], lambda_=lambda_)
        cpi_smooth2[:T + i - 1] = firstsmooth(y=cpi_smooth1[:T + i - 1], lambda_=lambda_)
        cpi_forecast[i - 1] = (2 + (lambda_ / (1 - lambda_))) * cpi_smooth1[T + i - 2] - \
                              (1 + (lambda_ / (1 - lambda_))) * cpi_smooth2[T + i - 2]
        #cpi_hat = 2 * cpi_smooth1[:T + i - 1] - cpi_smooth2[:T + i - 1]
        #sig_est = np.sqrt(np.var(cpi_dates.iloc[1:T + i - 1] - cpi_hat[:-1]))
        #cl[i - 1] = norm.ppf(1 - alpha_lev / 2) * sig_est
    return cpi_forecast
```

## Holt Winters (tercer orden)

```python
def tes_optimizer(train, val, trend, seasonal):
    
    alphas = betas = gammas = np.arange(0.1, 0.8, 0.1)
    abg = list(itertools.product(alphas, betas, gammas))
    best_alpha, best_beta, best_gamma, best_mape = None, None, None, 900
    
    for comb in abg:
        for seasonal_periods in range(2,7):
            tes_model = ExponentialSmoothing(train, trend=trend, seasonal=seasonal, seasonal_periods=seasonal_periods).\
                fit(smoothing_level=comb[0], smoothing_slope=comb[1], smoothing_seasonal=comb[2])
            y_pred = tes_model.forecast(len(val))
            mape = 100*(sum(abs(((val - y_pred)/val)))/len(val))
            if mape < best_mape:
                best_alpha, best_beta, best_gamma, best_mape, seasonal_periods = comb[0], comb[1], comb[2], mape, seasonal_periods
                
    return best_alpha, best_beta, best_gamma, best_mape, seasonal_periods
```

## assumptions

```python
def assumptions(fit):
    residuals=fit.resid
    a=0
    warning = ""
    
    #normality test
    #H0: The residuals have a normal distribution.
    #H1: The residuals don't have a normal distribution.
    print("prueba normalidad queremos que p>0.05")
    jb_stat, jb_pvalue, skewness, kurtosis = jarque_bera(residuals)
    print("p normalidad:",jb_pvalue)
    if jb_pvalue < 0.01:
        stat, shapiro_p = shapiro(residuals)
        print("p normalidad:",shapiro_p)
        if shapiro_p < 0.01:
            stat, p = kstest(residuals, 'norm')
            print("p normalidad:",p)
            if p<0.01:
                a=a+1
                warning = warning + " The residuals do not follow a normal distribution"
       
    #homocedasticity
    #H0: The residuals are homocedastic.
    #H1: The residuals aren't homocedastic.
    print("prueba homocedassticidad queremos que p>0.05")
    arch_stat, arch_pvalue, _, _ = het_arch(residuals)
    print("p heterocedasticidad:",arch_pvalue)
    if arch_pvalue < 0.01:
        warning= warning + " The residuals aren't homocedastic"
        a=a+1
        
    #independency
    #H0: The residuals are independent.
    #H1: The residuals aren't independent.    
    ljung_box_test = acorr_ljungbox(residuals, lags=[1], return_df=True)
    ljung_box_pvalue = ljung_box_test['lb_pvalue'].values[0]
    print("prueba independencia queremos que p>0.05")
    print("p independencia:",ljung_box_pvalue)
    if ljung_box_pvalue < 0.01:
        warning= warning + " The residuals aren't independent"
        a=a+1
        
    if a==0:
        warning = " The assumptions made about the residuals are verified with a confidence of " + str(1-0.05) + "."
                    
    warning = warning + " with a confidence of 0.99"
    
    return warning
```

## Moving blocks

```python
import numpy as np

def moving_block_bootstrap(data, block_size):
    """
    Moving Blocks Bootstrap for time series data.
    
    Parameters:
    data (array-like): The time series data to resample.
    block_size (int): The size of the blocks to resample.
    
    Returns:
    array: A bootstrapped time series sample.
    """
    n = len(data)
    blocks = [data[i:i + block_size] for i in range(n - block_size + 1)]
    
    # Resample blocks with replacement
    num_blocks = int(np.ceil(n / block_size))
    sampled_blocks = np.random.choice(range(len(blocks)), size=num_blocks, replace=True)
    
    # Concatenate sampled blocks to form a new time series
    bootstrapped_data = np.concatenate([blocks[i] for i in sampled_blocks], axis=0)
    
    return bootstrapped_data[:n]  # Ensure the length matches the original data
```
