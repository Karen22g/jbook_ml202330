#### 1. Merge weekly y monthly

```python
def sample_dates(n,df):
    
    """
    
    Groups the weekly data depending on the indicators' temporality into monthly or annual data.
    
    Args:
        n (Integer): Temporality needed: 1 if month, 2 if year.
        df (Dataframe): Weekly data
        
    Returns:
        df (Dataframe): Weekly information grouped per month or year.
        
    """
    
    #It is necessary to set the index to "date" to use the "resample code"
    df['date'] = pd.to_datetime(df['date'])
    df = df.set_index('date')
    
    if n==1:
        
        #Groups the data for each month of each year and calculates the mean
        df = df.resample('M').mean()
        end_dates = df.index
        new_dates = []
        
        #Matches the order of the weekly "date" column as is presented in the column "date" of the monthly_data dataframe, 
        #e.g. "25-01-2007" to "01-01-2007" to support the next merge process
        for k in end_dates:
            k=datetime.strptime(str(k), "%Y-%m-%d %H:%M:%S")
            k=k.strftime("%Y-%m-%d")
            date = datetime.strptime(k, "%Y-%m-%d")
            first_day = datetime(date.year, date.month, 1)
            new_dates.append(first_day.strftime("%Y-%m-%d"))
        
        #Creates a date columns that is used after to group the weekly and monthly information
        df['date']=new_dates
    
    if n==2:
        
        #Groups the data for each year and calculates the mean
        df =df.resample('Y').mean()
        end_dates = df.index
        new_dates = []
        
        #Takes the year of each observation
        for k in end_dates:
            k=datetime.strptime(str(k), '%Y-%m-%d %H:%M:%S')
            k=k.year
            new_dates.append(k)
            
        #Creates a date columns that is used after to group the weekly and annual information
        df['date']=new_dates
        
    return df
```

```python
def merge_date(indicator,n,dfx,dfy):
    
    """
    
    Joins the weekly data with the indicator of analysis.
    
    Args:
        Indicator (String): Indicator whose VAR Model is being created.
        n (Integer): Temporality needed: 1 if month, 2 if year.
        dfx (Dataframe): Weekly data
        dfy (Dataframe): Monthly or annual data
        
    Returns:
        dfx (Dataframe): Dataframe that contains the independent and dependent indicators.
    
    """
    
    dfx=sample_dates(n,dfx) #Calls the "sample_dates" function
    dfx.reset_index(drop=True, inplace=True)
    dfx=dfx.merge(dfy[['date',indicator]].dropna(), how='left', on='date').sort_values(by=['date'])
    dfx=dfx.dropna().reset_index(drop=True, inplace=False)
    #dfx=dfx.drop(['date'],axis=1)
    
    return dfx
```

#### 2. SDL (Seasonal decompose)

```python
def combinations(P_,Q_,p_,q_):
    
    """
    
    Generates all possible combinations of the p,q,P,Q parameters, that are used to create the (S)Arimax model.
    
    Args:
        P_ (Integer): Seasonal autorregresive parameter.
        Q_ (Integer): Seasonal moving average parameter.
        p_ (Integer): Autorregresive parameter of the indicator's timeseries.
        q_ (Integer): Moving average parameter of the indicator's timeseries.
        
    Returns:
        general (Array): Array with the different combinations generated, has arrays inside, each array is a unique combination of the parameters.
    
    """
        
    general=[]

    for p in range(p_):
        for q in range(q_):
                for P in range(P_):
                    for Q in range(Q_):
                        combinations=[]
                        combinations.append(p+1)
                        combinations.append(q+1)
                        combinations.append(P+1)
                        combinations.append(Q+1)
                        general.append(combinations)
                        
    return general
```

```python
def create_model(s,p,q,P,Q,d,D,indicator,dftraining):
    
    """
    
    Creates, filters, test and chooses the best (S)Arimax model. The parameter of selection is the AIC and the error's assumptions performance.
    
    Args:
        s (Bool): True if there's a seasonal component, false if there isn't.
        p (Integer): Autorregresive parameter of the indicator's timeseries.
        q (Integer): Moving average parameter of the indicator's timeseries.
        P (Integer): Seasonal autorregresive parameter.
        Q (Integer): Seasonal moving average parameter.
        d (Integer): Number of times that the indicator's timeseries was differentiated.
        D (Integer): Number of times that the seasonal component's timeseries was differentiated.
        indicator (String): Monthly indicator whose (S)Arimax model is being created.
        dftraining (Dataframe): Dataframe with values used to create and train the model, it has stationary values if a differentation was needed.

    Returns:
         a (Integer): Type of result given by the function (0 if the model selected passed the error's assumptions, 1 if it doesn't, 2 if none of the weekly indicator
                      was significative in the model for each of the combinations used).
         variables_sign (Array): Array with the exogenous variables (weekly indicators) that were significative.
         model ((S)Arimax fit): Fit model selected by the function.
         warning (String): Result of the error's assumptions test.
    
    """
    
    comb=combinations(P,Q,p,q)
    aics=[]
    fits=[]
    varsign=[]
    warnings=[]
    aics_all=[]
    varsign_all=[]
    fits_all=[]
    warnings_all=[]
    warning1=""
    
    #generate all possible combinations and choose by AIC
    
    for combination in comb:
        p=combination[0]
        q=combination[1]
        P=combination[2]
        Q=combination[3]
        
        if s==True: #Seasonal component, we use SARIMAX

            model=SARIMAX(dftraining[indicator], exog=dftraining[weekly],
                               order=(p, d, q),
                               seasonal_order=(P,D,Q,12),
                               mle_regression=True).fit()

            pvalues=model.pvalues[:18]
            columns=pvalues.index[:18]
            variables_sign=[]
            i=0

            for pv in pvalues:
                if pv<0.05:
                    variables_sign.append(columns[i])
                i=i+1
                
            if len(variables_sign)==0:
                warning="There aren't significant weekly indicators for this model"
                continue
            
            if len(variables_sign)>0:
                
                model=SARIMAX(dftraining[indicator], exog=dftraining[variables_sign],
                                       order=(p, d, q),
                                       seasonal_order=(P,D,Q,12),
                                       mle_regression=True).fit()

            a,warning=assumptions(model)
            warning=warning1+warning

        if s==False: #No seasonal component, we use ARIMAX

            model=ARIMA(dftraining[indicator], exog=dftraining[weekly],
                            order=(p, d, q)).fit(method='nm')

            pvalues=model.pvalues[:18]
            columns=pvalues.index[:18]
            variables_sign=[]
            i=0

            for pv in pvalues:
                if pv<0.05:
                    variables_sign.append(columns[i])
                i=i+1
                
            if len(variables_sign)==0:
                warning="There aren't significant weekly indicators for this model"
                continue
            
            if len(variables_sign)>0:
                model=ARIMA(dftraining[indicator], exog=dftraining[variables_sign],order=(p, d, q)).fit()

            a,warning=assumptions(model)
    
        print(a)
        if a==0: #The models that passed the error's assumptions, also the exogenous variables used in that case

            fits.append(model)
            aics.append(model.aic)
            varsign.append(variables_sign)
            warnings.append(warning)
            continue
        
        #All models are stored, also the exogenous variables used in that case

        fits_all.append(model)
        aics_all.append(model.aic)
        varsign_all.append(variables_sign)
        warnings_all.append(warning)
        
    #In this steps, we're going to choose the best model best based on the minimum AIC value
        
    minaic=99999999
    
    if len(variables_sign)==0:
        a=2
    
    if len(variables_sign)>0:
        
        if len(aics)>0:
            a=0
        #Here we take the models that passed the error's assumptions with the best minimum AIC
            for AIC in aics:                
                if AIC<minaic:
                    minaic=AIC
                    pos=aics.index(AIC)
                    model=fits[pos]
                    warning=warnings[pos]
                    variables_sign=varsign[pos]
         
        #Here we have the models that didn't pass the error's assumptions, also we choose it with the best minimum AIC
        if len(aics)==0:
            a=1
            for AIC in aics_all:
                if AIC<minaic:
                    minaic=AIC
                    pos=aics_all.index(AIC)
                    model=fits_all[pos]
                    warning=warnings_all[pos]
                    variables_sign=varsign_all[pos]

    return a,variables_sign,model,warning
```

```python
def df_summaries(model,indicator,varsign):
    
    """
    
    Generates the summary for each model created, complete or just with the significant weekly indicators.
    
    Args:
        indicator (String): Monthly indicator whose (S)Arimax model is being created.
        model ((S)Arimax fit): Fit model selected for the indicator given.
        varsign (Array): Array with the exogenous variables (weekly indicators) that were significative.

    Returns:
         dfc (Dataframe): Dataframe with all the variables and parameters used in the model.
         df (Dataframe): Dataframe with just the coefficients and result of the weekly indicators that were significative.
    
    """
    
    #complete dataframe
    dfc=pd.DataFrame({'Variable':model.params.index,'Coefficients':model.params}).reset_index(drop=True)
    dfc['Indicator']=indicator

    #just significant variables dataframe
    df=pd.DataFrame({'Variable':model.params[:len(varsign)].index,'Coefficients':model.params[:len(varsign)]}).reset_index(drop=True)
    df['Indicator']=indicator
    
    return dfc,df
```

#### 3. Getting d and D for SARIMAX components

```python
def stationarity(df):
    
    """
    
    Test the stationarity of the timeseries contained in the dataframe and if it is false, then differentiates the dataframe
    until all the timeseries are stationary.
    
    Args:
        df (Dataframe): Dataframe that contains both the dependent and independent indicators.
   
    Returns:
        df (Dataframe): Dataframe that contains the independent and dependent indicators but stationaries.
        d (Integer): Number of times that the dataframe was differentiated.
    
    """
    
    st=1
    d=0
    
    while st>0:
        columns=[]
        for col in df.columns.tolist():
            #ADF Test for stationarity:
            #H0: The timeseries is not stationary
            #H1: The timeseries is stationary
            r=adfuller(df[col], autolag='AIC')
            output = {'test_statistic':round(r[0], 4), 'pvalue':round(r[1], 4), 'n_lags':round(r[2], 4), 'n_obs':r[3]}
            n_lags = output['n_lags']
            p_value = output['pvalue']
            if p_value>=0.05:
                columns.append(col) #We keep the column that makes necessary the differentiation process
                
        st=len(columns)
        if st>0:
            d=d+1
            if d==3: #Notice that we consider inappropriate to differentiate the dataframe more than 2 times
                print("WARNING: The next variable(s) aren't stationary after a second differentation process:", columns)
            df=df.diff()
            df=df.dropna()
            
    return d,df
```

#### 6. Error's assumptions

```python
def assumptions(fit):
    
    """
    
    Test the errors' assumptions for the Model created.
    
    Args:
        fit (Model): Model to test.
        
    Returns:
    
        a (Integer): Number of tests that the residuals don't pass.
        warning (String): Description of the errors' assumptions tests.
        
    """

    residuals=fit.resid
    
    ci=[0.05,0.025,0.01]
    
    for pv in ci:
        a=0
        warning=""
        
        #homocedasticity
        #H0: The residuals are homocedastic.
        #H1: The residuals aren't homocedastic.
        
        white_test = het_white(residuals,  sm.add_constant(fit.model.exog))
        
        if white_test[1]<pv:
            
            resid1 = residuals[:int(len(residuals)*0.5)]
            resid2 = residuals[int(len(residuals)*0.5):]
            res = stats.fligner(resid1, resid2)
            p = res.pvalue
            
            if p<pv:
                warning=warning + " The residuals aren't homocedastic"
                a=a+1


        #normality test
        #H0: The residuals have a normal distribution.
        #H1: The residuals don't have a normal distribution.
        stat, p = shapiro(residuals)

        if p<pv:
            stat, p = kstest(residuals, 'norm')

            if p<pv:
                a=a+1
                warning = warning + " The residuals do not follow a normal distribution"

        #independency
        #H0: The residuals are independent.
        #H1: The residuals aren't independent.
        out = durbin_watson(residuals)
        if out < 1.5 and out >=2.5:
            a=a+1
            warning = warning + " The residuals aren't independent"
        
        warning = warning + " in none of the confidence levels."

        if a==0:
            warning = "The assumptions made about the residuals are verified with a confidence of " + str(1-pv) + "."
            break
    
    return a,warning
```

#### 7. Forecasting the testing

```python
def model_forecasting(indicator,varsign,d,dftest,dftraining,fit,dftrain):
    
    """
    
    Generates the values forecasted by the model and then turns them into the original values without being differentiated.
    
    Args:
    
        indicator (String): Monthly indicator whose (S)Arimax model is being created.
        varsign (Array): Array with the exogenous variables (weekly indicators) that were significative.
        d (Integer): Number of times that the dataframe was differentiated.
        dftest (Dataframe): Testing original values (without differentation).
        dftraining (Dataframe): Training values used to create the (S)Arimax model.
        fit ((S)Arimax fit): Fit model selected for the indicator given.
        dftrain (Dataframe): Training original values (without differentation), these values are used to return the differentiated values to the original scale.
    
    Returns:
    
        dftest (Dataframe): Testing original values (without differentation) plus the forecasted values of the indicator given.
        mape (float): Mean Absolute Percentage Error
        mae (float): Mean Absolute Error
        rmse (float): Root mean Square Error
        
    """
    
    startp=dftraining.shape[0]
    endp=startp+dftest.shape[0]-1
    dftest['forecast']=fit.predict(start=startp,end=endp,dynamic=True, exog=dftraining.tail(5)[varsign],
                                     return_conf_int=True)

    #To get the original values
    if d==2:
        dftest['forecast']=dftrain[indicator].iloc[-1]-dftrain[indicator].iloc[-2] + dftest['forecast'].cumsum()
        d=d-1
    if d==1:
        dftest['forecast']=dftrain[indicator].iloc[-1] + dftest['forecast'].cumsum()

    mape,mae,rmse=forecast_accuracy(dftest['forecast'], dftest[indicator])
    
    return dftest,mape,mae,rmse
```

```python
def forecast_accuracy(forecast, actual):
    
    """
    
    Calculates the forecasting's error accuracy.
    
    Args:
        forecast (Dataframe): Indicator's testing values predicted by the model.
        actual (Dataframe): Indicator's original testing values.
        
    Returns:
    
        mape (float): Mean Absolute Percentage Error
        mae (float): Mean Absolute Error
        rmse (float): Root mean Square Error
        
    """

    mape = np.mean(np.abs(forecast - actual)/np.abs(actual))  # MAPE
    mae = np.mean(np.abs(forecast - actual))    # MAE
    rmse = np.mean((forecast - actual)**2)**.5  # RMSE
    return mape, mae, rmse
```
